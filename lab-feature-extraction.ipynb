```python
# Rename the columns so they follow the PE8 (snake case).
customer_df.columns = customer_df.columns.str.lower().str.replace(" ","_")
customer_df.rename(columns = {"employmentstatus" : "employment_status"}, inplace = True)
customer_df.columns
```

#### 4. Plot time variable. Can you extract something from it?

```python
fig, ax = plt.subplots(figsize=(12, 5))

sns.lineplot(customer_df['effective_to_date'], customer_df['total_claim_amount'], ci=None)

ax.set_title('Total claim amount over time', fontsize = 20, loc='center')
ax.set_xlabel('Date', fontsize = 15)
ax.set_ylabel('Total claim amount', fontsize = 15)
```

```python
# Creating a copy of the dataset in order to plot the time variable.
plot_time = customer_df.copy()
```

```python
# Creating 3 new columns: day, month and year of the effective_to_date.
plot_time['day']= plot_time['effective_to_date'].dt.day
plot_time['month']= plot_time['effective_to_date'].dt.month
plot_time['year']= plot_time['effective_to_date'].dt.year
plot_time.head()
```

```python
# There is only one year(2011)
plot_time['year'].value_counts()
```

```python
# There are only two months.
plot_time['month'].value_counts()
```

```python
fig, ax = plt.subplots(figsize=(12, 5))

palette = sns.color_palette("Set2", 2)

sns.lineplot(data=plot_time, x=plot_time['day'], y=plot_time['total_claim_amount'], hue=plot_time['month'], palette=palette, ci=None)
ax.set_title('Total claim amount in months', fontsize = 20, loc='center')
ax.set_xlabel('Day', fontsize = 15)
ax.set_ylabel('Total claim amount', fontsize = 15)
```

* There seems to be a general monthly trend.
